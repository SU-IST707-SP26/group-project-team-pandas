### Supporting Notebook Index

This project is organized into the following notebooks in the `work` folder:

- `01.2_data_loading_and_cleaning.ipynb`  
  Loads and cleans the NYC restaurant inspection data, including filtering to 2022–2025.

- `02_exploratory_data_analysis.ipynb`  
  Explores inspection trends, score distributions, grades, and borough patterns.

- `03_inspection_level_preprocessing.ipynb`  
  Transforms violation-level data into inspection-level records and prepares data for feature engineering.

- `04_feature_engineering.ipynb`  
  Creates historical, temporal, and target features, including the inspection likelihood variable.

- `05_baseline_modeling.ipynb`  
  Builds and compares logistic regression, balanced logistic regression, and random forest models, including tuning.

- `06_model_evaluation_and_interpretation.ipynb`  
  Evaluates model performance and interprets results using feature importance and confusion matrix analysis.

- `07_visualizations_for_report.ipynb`  
  Generates final visualizations used in the report and presentation.

- `wholecode.ipynb`  
  Contains the initial full end-to-end workflow used during development.
