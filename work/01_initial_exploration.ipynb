```python
#get rid of nulls
def clean_restaurant_inspection_data(df):
    # Make a copy to avoid modifying original
    df_clean = df.copy()
    
    # Basic info cleaning
    df_clean['DBA'] = df_clean['DBA'].fillna("Unknown Business")
    df_clean['BUILDING'] = df_clean['BUILDING'].fillna("No Number")
    df_clean['STREET'] = df_clean['STREET'].fillna("Unknown Street")
    df_clean['PHONE'] = df_clean['PHONE'].fillna("No Phone Listed")
    
    # Zipcode by borough
    df_clean['ZIPCODE'] = df_clean.groupby('BORO')['ZIPCODE'].transform(lambda x: x.fillna(x.median()))
    
    # Inspection related
    df_clean['CUISINE DESCRIPTION'] = df_clean['CUISINE DESCRIPTION'].fillna("Unspecified")
    df_clean['ACTION'] = df_clean['ACTION'].fillna("No Action Recorded")
    df_clean['INSPECTION TYPE'] = df_clean['INSPECTION TYPE'].fillna("Unspecified")
    df_clean['VIOLATION CODE'] = df_clean['VIOLATION CODE'].fillna("No Violation")
    df_clean['VIOLATION DESCRIPTION'] = df_clean['VIOLATION DESCRIPTION'].fillna("No Violation Found")
    
    # Grades and scores
    df_clean['SCORE'] = df_clean['SCORE'].fillna(-1)
    df_clean['GRADE'] = df_clean['GRADE'].fillna("Not Graded")
    df_clean['GRADE DATE'] = df_clean['GRADE DATE'].fillna(df_clean['INSPECTION DATE'])
    
    # Location data
    for coord in ['Latitude', 'Longitude']:
        df_clean[coord] = df_clean.groupby('BORO')[coord].transform(lambda x: x.fillna(x.median()))
    
    # Administrative boundaries - FIX HERE
    admin_cols = ['Community Board', 'Council District', 'Census Tract', 'NTA']
    for col in admin_cols:
        df_clean[col] = df_clean.groupby('BORO')[col].transform(lambda x: 
            x.fillna(x.mode().iloc[0] if not x.mode().empty else -1))
    
    # Building numbers
    df_clean['BIN'] = df_clean['BIN'].fillna(0)
    df_clean['BBL'] = df_clean['BBL'].fillna(0)
    
    # Location string
    df_clean['Location'] = df_clean['Location'].fillna('Unknown Location')
    
    return df_clean
```

```python
# Import Required Libraries
!pip install matplotlib
!pip install seaborn

import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

# Set display options
pd.set_option('display.max_columns', None)
pd.set_option('display.max_colwidth', None)
```

```python
# Convert INSPECTION DATE to datetime
df["INSPECTION DATE"] = pd.to_datetime(df["INSPECTION DATE"], errors="coerce")

# Filter data between 2015 and 2025
df_2015_2025 = df[
    (df["INSPECTION DATE"].dt.year >= 2015) &
    (df["INSPECTION DATE"].dt.year <= 2025)
].copy()
```

```python
print("Original:", df.shape)
print("2015–2025:", df_2015_2025.shape)

print(
    df_2015_2025["INSPECTION DATE"].min(),
    "to",
    df_2015_2025["INSPECTION DATE"].max()
)
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

plt.figure(figsize=(6,4))
sns.countplot(data=df_2015_2025, x="GRADE", order=["A","B","C"])

plt.title("Distribution of Inspection Grades (2015–2025)")
plt.xlabel("Grade")
plt.ylabel("Count")

plt.show()
```

```python
plt.figure(figsize=(8,5))

sns.countplot(
    data=df_2015_2025,
    x="BORO",
    hue="GRADE",
    order=df_2015_2025["BORO"].value_counts().index
)

plt.title("Inspection Grades by Borough")
plt.xlabel("Borough")
plt.ylabel("Count")
plt.legend(title="Grade")

plt.show()
```

```python
df_2015_2025["YearMonth"] = df_2015_2025["INSPECTION DATE"].dt.to_period("M")

monthly = df_2015_2025.groupby("YearMonth")["CAMIS"].count()

plt.figure(figsize=(12,5))
monthly.plot()

plt.title("Inspection Volume Over Time (2015–2025)")
plt.xlabel("Year")
plt.ylabel("Number of Inspections")

plt.show()
```

```python
plt.figure(figsize=(6,4))

sns.boxplot(
    data=df_2015_2025,
    x="GRADE",
    y="SCORE",
    order=["A","B","C"]
)

plt.title("Inspection Scores by Grade")
plt.xlabel("Grade")
plt.ylabel("Score")

plt.show()
```
